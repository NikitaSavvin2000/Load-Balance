```python
# Statistical Tests:formal statistical tests to assess normality. Common tests include the Shapiro-Wilk test and the Anderson-Darling test. These tests provide a p-value, and a low p-value indicates that your data significantly deviates from a Gaussian distribution.

import pandas as pd
from scipy.stats import shapiro, anderson

# Load your data from the CSV file into a pandas DataFrame
data = pd.read_csv('result/P_l.csv')

# Specify the column you want to analyze for normality
column_name = 'P_l'  # Replace with the actual column name

# Extract the data from the specified column
data_to_test = data[column_name]

# Shapiro-Wilk test
stat, p = shapiro(data_to_test)
print(f'Shapiro-Wilk Test: Statistic={stat}, p-value={p}')

# Anderson-Darling test
result = anderson(data_to_test)
print(f'Anderson-Darling Test: Statistic={result.statistic}, Critical Values={result.critical_values}, Significance Levels={result.significance_level}')
```

```python
# Histogram:

import matplotlib.pyplot as plt

# Assuming 'data' is your dataset and 'column_name' is the column you want to visualize
data = pd.read_csv('result/P_l.csv')
column_name = 'P_l'  # Replace with the actual column name

# Extract the data from the specified column
data_to_plot = data[column_name]

# Create the histogram for the selected column
plt.hist(data_to_plot, bins=20)  # Adjust the number of bins as needed
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')
plt.show()

```

```python
# Box Plot:

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your dataset and 'column_name' is the column you want to visualize
data = pd.read_csv('result/P_l.csv')
column_name = 'P_l'  # Replace with the actual column name

# Extract the data from the specified column
data_to_plot = data[column_name]

# Create the box plot for the selected column
sns.boxplot(x=data_to_plot)
plt.xlabel('Value')
plt.title('Box Plot of P_l')
plt.show()
```

```python
# Kernel Density Estimate (KDE) Plot:

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your dataset and 'column_name' is the column you want to visualize
data = pd.read_csv('result/P_l.csv')
column_name = 'P_l'  # Replace with the actual column name

# Extract the data from the specified column
data_to_plot = data[column_name]

# Create the KDE plot for the selected column
sns.kdeplot(data_to_plot, shade=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Kernel Density Estimate (KDE) Plot')
plt.show()
```

```python
# Quantile-Quantile (Q-Q) Plot:

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load your data from the CSV file into a pandas DataFrame
data = pd.read_csv('result/P_l.csv')

# Extract the column of interest from the DataFrame, assuming it's called 'column_name'
column_name = 'P_l'
data_to_plot = data[column_name]

# Generate the Q-Q plot
sm.qqplot(data_to_plot, line='s')
plt.show()
```
